```python
import os 
from nltk.corpus import stopwords
import numpy as np
import pandas as pd
```

```python
#getting all the categories
all_categories = os.listdir("20_newsgroups")
print(len(all_categories))
all_categories
```

```python
#creating a set of all the stopwords in english
stop_words = set(stopwords.words("english"))
stop_words
```

```python
def count_word_freq(all_categories):
    word_freq_count = {}
    n_documents = 0
    for category in all_categories:
        all_documents = os.listdir("20_newsgroups/"+category)
        all_documents_train = all_documents[0:750]
        n_documents += len(all_documents_train)
        for document in all_documents_train:
            file_words = open("20_newsgroups/"+category+"/"+document).read().replace("\n", " ").split(" ")
            for word in file_words:
                if(word not in stop_words):
                    if(word not in word_freq_count.keys()):
                        word_freq_count[word] = 1
                    else:
                        word_freq_count[word] += 1
    return word_freq_count, n_documents        
```

```python
word_freq_count = count_word_freq(all_categories)[0]
word_freq_count
```

```python
def get_fetaure_words(word_freq_count, n):
    feature_words_freq = {}
    for word in word_freq_count.keys():
        if(word_freq_count[word]>n):
            feature_words_freq[word] = word_freq_count[word]
    return feature_words_freq
feature_words_freq = get_fetaure_words(word_freq_count, 50)
feature_words_freq
```

```python
vocabulary = feature_words_freq.keys()
print(len(vocabulary))
vocabulary
```

```python
n_documents = count_word_freq(all_categories)[1]
word_count_df = pd.DataFrame(data = np.zeros((n_documents, len(vocabulary))), columns = vocabulary)
```

```python
y_train = []
doc_no = 0
for category in all_categories:
    all_documents = os.listdir("20_newsgroups/"+category)
    all_documents_train = all_documents[0:750]    
    for document in all_documents_train:
        y_train.append(category)
        file_words = open("20_newsgroups/"+category+"/"+document).read().replace("\n", " ").split(" ")
        for word in file_words:
            if(word in vocabulary):
                word_count_df[word][doc_no] += 1
        doc_no += 1
```

```python
word_count_df
```

```python
y_train = np.array(y_train)
x_train = word_count_df.values
```

```python
x_train.shape
```

```python
category_word_count = {}
for category in all_categories:
    category_word_count[category] = {}  
    category_word_count[category]["total_words"] = 0
    for word in vocabulary:
        category_word_count[category][word] = word_count_df[y_train == category][word].sum()
        category_word_count[category]["total_words"] += word_count_df[y_train == category][word].sum()
```

```python
category_word_count
```

```python
y_pred = []
for category in all_categories:
    all_documents = os.listdir("20_newsgroups/"+category)
    all_documents_test = all_documents[750:]    
    for document in all_documents_test:
        max_pDoc = -10000
        file_words = open("20_newsgroups/"+category+"/"+document).read().replace("\n", " ").split(" ")
        for d_category in category_word_count.keys():
            for word in file_words:
                if(word in vocabulary):                
                    pDocument = np.log(len(y_train[y_train == d_category])/len(y_train))
                    p_word = np.log((category_word_count[d_category][word]+1)/(category_word_count[d_category]["total_words"] + len(vocabulary)))
                    pDocument += p_word
            if(max_pDoc <= pDocument):
                max_pDoc = pDocument
                doc_category = category
        y_pred.append(doc_category)
```
