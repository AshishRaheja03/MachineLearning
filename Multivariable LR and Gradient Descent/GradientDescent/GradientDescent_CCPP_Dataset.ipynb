```python
import numpy as np 
import pandas as pd
```

```python
ccpp_data_train = pd.read_csv("CCPP_train.csv")     #loading data as DataFrame using read_csv() from pandas 
ccpp_df_train = ccpp_data_train.copy()       #copying ccpp_data to another dataframe
ccpp_df_train.head()
```

```python
ccpp_df_train.describe()
```

```python
ccpp_train = np.array(ccpp_df_train)      #converting dataframe to numpy array
X_train = ccpp_train[:,0:4]
Y_train = ccpp_train[:,4]
print(X_train.shape)
print(Y_train.shape)
```

```python
x_train_df = pd.DataFrame(X_train)
x_train_df["last_col"] = 1
X_train = x_train_df.values
X_train.shape
```

```python
def step_GD(x, y, lr, m):
    m_slope_arr = []
    for k in range(x.shape[1]):        
        slope = 0
        for i in range(x.shape[0]):
            mx = 0
            for j in range(x.shape[1]):
                mx += m[j]*x[i][j]
            slope += (-2/x.shape[0])*(y[i] - mx)*x[i][k]
        m_slope_arr.append(slope)
    m_slope = np.array(m_slope_arr)
    new_m = m - lr*m_slope 
    return new_m
```

```python
def cost(x, y, m):
    tcost = 0
    for i in range(x.shape[0]):
        mx = 0
        for j in range(x.shape[1]):
            mx += m[j]*x[i][j]
        tcost += ((y[i] - mx)**2)/x.shape[0]
    return tcost
```

```python
def GradientDescent(x, y, lr, num_itr):
    m = np.zeros(x.shape[1])
    for i in range(num_itr):
        m = step_GD(x, y, lr, m)
        print("cost = ", cost(x, y, m))
    return m
```

```python
m = GradientDescent(X_train, Y_train,0.0000001,100000)
print("m = ", m)
```

```python
x_test_df = pd.read_csv("CCPP_test.csv")      #loading data as DataFrame using read_csv() from pandas 
x_test_df["last"] = 1
x_test = x_test_df.values
```

```python
y_test_pred = []
for i in range(x_test.shape[0]):
    y_temp = 0
    for j in range(len(m)):
        y_temp += m[j]*x_test[i][j]
    y_test_pred.append(y_temp)
len(y_test_pred)
```

```python
np.savetxt("CCPP_Y_predicted.csv", y_test_pred, fmt = "%s", delimiter = ",")
```
