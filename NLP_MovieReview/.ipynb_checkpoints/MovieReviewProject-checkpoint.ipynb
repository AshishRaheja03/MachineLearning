```python
#get the data in the form [(words_in_files, category_of_files)]
documents = []
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(fileid), category))
documents
```

```python
#to get the shuffeled documents to get some randomness in files so that we can perform train test split 
import random
random.shuffle(documents)
documents
```

```python
from nltk.corpus import wordnet
#function to convert output of nltk.pos_tag to wordnet pos tags as lemmatizer used is WordNetLemmatizer
def get_simple_pos(pos):
    if pos.startswith("N"):
        return wordnet.NOUN
    elif pos.startswith("J"):
        return wordnet.ADJ
    elif pos.startswith("V"):
        return wordnet.VERB
    elif pos.startswith("R"):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk.corpus import stopwords           #contains stopwords
import string          #used to get punctuations
stopWords = set(stopwords.words("english"))
stopWords.update(list(string.punctuation))
stopWords         #list of stopwords including punctuations 
```

```python
from nltk import pos_tag      #to get part of speech of a word
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
def clean_words_review(words):
    cleaned_words = []
    for w in words:
        if w.lower() not in stopWords:
            pos = pos_tag([w])
            cleaned_words.append(lemmatizer.lemmatize(w, pos=get_simple_pos(pos[0][1])))
    return cleaned_words
```

```python
clean_W_documents = [(clean_words_review(words),category) for words, category in documents]
```

```python
clean_W_documents[0]
```

```python
#doing the training(75%) and testing(25%) split 
training_documents = clean_W_documents[:1500]
testing_documents = clean_W_documents[1500:]
```

```python
#getting all words to further choose feture among them based on their frequency in the documents
#feature selection is totlly based on training data
all_words = []
for doc in training_documents:
    all_words += doc[0]
all_words
```

```python
from nltk import FreqDist     #to get the frequency distribution of all words 
len(FreqDist(all_words))
```

```python
common_words = FreqDist(all_words).most_common(15000)      #getting the 15000 most common words
```

```python
feature_words = [i[0] for i in common_words]
len(feature_words)
```

```python
#getting the data ready in the format that is to be feeded to nltk classifier
#data format: [({words:count}, category)]   a tuple for each document
def feature_dict_create(words):
    category_features = {}
    for w in feature_words:
        category_features[w] = w in set(words) 
    return category_features
```

```python
training_data = [(feature_dict_create(words), category) for words, category in training_documents]
testing_data = [(feature_dict_create(words), category) for words, category in testing_documents]
```

```python
from nltk import NaiveBayesClassifier
NBclassifier = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(NBclassifier, testing_data)
```
