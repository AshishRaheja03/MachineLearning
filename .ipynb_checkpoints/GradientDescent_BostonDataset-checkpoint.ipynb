```python
import numpy as np
import pandas as pd
from sklearn import datasets, model_selection
```

```python
boston_train_data = np.loadtxt("Boston_train.csv", delimiter = ",")
boston_train_data.shape
```

```python
X_train = boston_train_data[:,0:-1]
Y_train = boston_train_data[:,-1]
print(X_train.shape)
Y_train.shape
```

```python
X_train_df = pd.DataFrame(X_train)
X_train_df.head()
```

```python
X_train_df["13"] = 1
X_train_df.head()
```

```python
X_train = X_train_df.values
X_train.shape
```

```python
def m_slopeCal(x, y, m):
    m_slope_arr = []
    for k in range(x.shape[1]):        
        slope = 0
        for i in range(x.shape[0]):
            mx = 0
            for j in range(x.shape[1]):
                mx += m[j]*x[i][j]
            slope += (-2/x.shape[0])*(y[i] - mx)*x[i][k]
        m_slope_arr.append(slope)
    m_slope = np.array(m_slope_arr)
    return m_slope
```

```python
def step_GD(x, y, lr, m):
    m_slope = m_slopeCal(x, y, m)
    new_m = m - lr*m_slope 
    return new_m
```

```python
def cost(x, y, m):
    tcost = 0
    for i in range(x.shape[0]):
        mx = 0
        for j in range(x.shape[1]):
            mx += m[j]*x[i][j]
        tcost += ((y[i] - mx)**2)/x.shape[0]
    return tcost
```

```python
def GradientDescent_self(x, y, lr, num_itr):
    m = np.zeros(x.shape[1])
    for i in range(num_itr):
        m = step_GD(x, y, lr, m)
        print("cost = ", cost(x, y, m))
    return m
```

```python
m = GradientDescent_self(X_train, Y_train, 0.005, 5000)
print("m = ",m)
```

```python
x_test = np.loadtxt("Boston_test.csv", delimiter = ",")
x_test.shape
```

```python
x_test_df = pd.DataFrame(x_test)
x_test_df.head()
```

```python
x_test_df["13"] = 1
x_test_df.head()
```

```python
x_test = x_test_df.values
x_test.shape
```

```python
y_test_pred = []
for i in range(x_test.shape[0]):
    y_temp = 0
    for j in range(len(m)):
        y_temp += m[j]*x_test[i][j]
    y_test_pred.append(y_temp)
len(y_test_pred)
```
